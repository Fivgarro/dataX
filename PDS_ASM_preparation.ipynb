```python
# Fill NA values with mean values
dtf = dtf.fillna(dtf.mean())
# Filter out rows with values >100 in 'age' column
dtf = dtf[dtf.Age <100]
# Filter out rows with NA values
dtf = dtf.dropna()
# Trim extra whitespaces
dtf = dtf.replace(r'\s*(.*?)\s*', r'\1', regex=True)
```

```python
# Detect and filter outliers
p95, p5 = np.percentile(dtf[['Age']], [95, 5])
print(p95, p5)
print(len(dtf))
dtf = dtf[dtf.Age < p95]
dtf = dtf[dtf.Age > p5]
print(len(dtf))
```
